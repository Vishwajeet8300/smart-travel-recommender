```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors

# Assuming pca_data already exists
# Step 1: Select numeric columns only
numeric_cols = pca_data.select_dtypes(include=['int64', 'float64']).columns
X = pca_data[numeric_cols]

# Step 2: Scale the numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_scaled)

# Step 4: Create a DataFrame with PCA results
pca_df = pd.DataFrame(pca_result, columns=["PC1", "PC2"])

# Step 5: Fit KNN on PCA results
knn_model = NearestNeighbors(n_neighbors=5, metric='euclidean')
knn_model.fit(pca_df)

# Step 6: Query similar destinations to the first one
distances, indices = knn_model.kneighbors([pca_df.iloc[0]])

# Step 7: Display results
print("🔍 Input destination:")
print(pca_data.iloc[0][['remainder__Name', 'remainder__City', 'remainder__State']])

print("\n✨ Top 5 similar destinations:")
print(pca_data.iloc[indices[0]][['remainder__Name', 'remainder__City', 'remainder__State']])
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Load the data (if not already)
data = pd.read_csv("final.csv")

# Drop unwanted columns
if "Unnamed: 0" in data.columns:
    data = data.drop(columns=["Unnamed: 0"])

# Define categorical columns to encode
categorical_cols = ["Zone", "Significance"]

# Fill missing values and ensure they're strings
data[categorical_cols] = data[categorical_cols].fillna("Unknown").astype(str)

# Create OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Build ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ("onehot", encoder, categorical_cols)
    ],
    remainder="passthrough"  # Keep all other columns as is
)

# Fit and transform the data
encoded_features = preprocessor.fit_transform(data)

# Get new feature names
encoded_feature_names = preprocessor.get_feature_names_out()

# Convert back to DataFrame
encoded_df = pd.DataFrame(encoded_features, columns=encoded_feature_names)

# Optional: reset index if needed
encoded_df.index = data.index  # Match index to original DataFrame

# Preview the final DataFrame
print(encoded_df.head())
```
